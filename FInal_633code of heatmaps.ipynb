```python
import io
import pandas as pd
import numpy as np
from random import random
from bs4 import BeautifulSoup
import requests
import plotly.graph_objects as go
import psycopg2
import matplotlib.pyplot as plt
import plotly
```

```python
df = pd.read_csv('/content/drive/MyDrive/1.csv')

for col in df.columns:
    df[col] = df[col].astype(str)

df['text'] = df['state'] 
fig = go.Figure(data=go.Choropleth(
    locations=df['state'],
    z=df['tot_cases'].astype(float),
    locationmode='USA-states',
    colorscale='Reds',
    autocolorscale=False,
    text=df['text'],
    marker_line_color='white',
    colorbar_title="Millions USD"
))

fig.update_layout(
    title_text='Covid-19 Cases in US<br>(Hover for breakdown)',
    geo = dict(
        scope='usa',
        projection=go.layout.geo.Projection(type = 'albers usa'),
        showlakes=True,
        lakecolor='rgb(255, 255, 255)'),
)

fig.show()
```

```python
import plotly.express as px

df = pd.read_csv('/content/drive/MyDrive/3.csv')
df

scl = [[0.0, '#ffffff'],[0.2, '#b4a8ce'],[0.4, '#8573a9'],
       [0.6, '#7159a3'],[0.8, '#5732a1'],[1.0, '#2c0579']] # purples

data_slider = []
for date in df['Date'].unique():
    df_segmented =  df[(df['Date']== date)]

    for col in df_segmented.columns:
        df_segmented[col] = df_segmented[col].astype(str)

    data_each_yr = dict(
                        type='choropleth',
                        locations = df_segmented['state'],
                        z=df_segmented['tot_cases'].astype(float),
                        locationmode='USA-states',
                        colorscale = 'Reds',
                        colorbar= {'title':'# Cases'})

    data_slider.append(data_each_yr)

steps = []
for i in range(len(data_slider)):
    step = dict(method='restyle',
                args=['visible', [False] * len(data_slider)],
                label='Day {}'.format(i + 1))
    step['args'][1][i] = True
    steps.append(step)

sliders = [dict(active=0, pad={"t": 1}, steps=steps)]

layout = dict(title ='......', geo=dict(scope='usa',
                       projection={'type': 'albers usa'}),
              sliders=sliders)

fig = dict(data=data_slider, layout=layout)
plotly.offline.iplot(fig)
```
